```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('../data/housing_data.csv')
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Summary statistics
print(data.describe())
```

```python
# Histogram of the target variable (Price)
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Price', kde=True)
plt.title('Distribution of Housing Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation heatmap
corr = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot of features vs. target variable
plt.figure(figsize=(12, 8))
sns.pairplot(data, x_vars=['Size (sqft)', 'Bedrooms', 'Bathrooms', 'Garage'], y_vars='Price', height=3, aspect=1, kind='scatter')
plt.suptitle('Scatter Plot of Features vs. Price')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
